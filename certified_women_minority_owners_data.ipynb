```python
## dropping columns with less relevant information 
df = df.drop(["MailingAddress1","MailingAddress2","MailingCity","MailingState","MailingZip", "ZIP","Goods_Materials_Suplier","Types_of_Construction_Projects", "fax", "Address1", "Address2", "Job_Exp4_Name_of_Client","Job_Exp4_Value_of_Contract","Job_Exp4_Percent_Self_Performed", "DATEOFJOB4","EXPERIENCE4", "Job_Exp2_Name_of_Client","Job_Exp2_Value_of_Contract","Job_Exp2_Percent_Self_Performed","DATEOFJOB2","Job_Exp3_Name_of_Client","Job_Exp3_Value_of_Contract","Job_Exp3_Percent_Self_Performed","DATEOFJOB3","EXPERIENCE3"], axis = 1)
df
```

```python
## further removing irrelevant columns 
df = df.drop(["Website", "Aggregate_Bonding_Limit", "Job_Exp1_Name_of_Client", "Job_Exp1_Value_of_Contract","Job_Exp1_Percent_Self_Performed","Vendor_DBA",], axis = 1)
df
```

```python
## renaming columns 
df = df.rename(columns = {"Vendor_Formal_Name": "vendor_formal_name",
                         "Contact_Name": "contact_name",
                         "Cert_Renewal_Date": "cert_renewal_date",
                         "Ethnicity": "ethnicity",
                         "City": "city",
                          "Brooklyn": "brooklyn",
                          "Bronx": "bronx"
                         "State": "state",
                         "Date_Of_Establishment": "date_of_establishment",
                         "Business_Description": "business_description",
                         "Signatory_To_Union_Contracts": "signatory_to_union_contracts",
                         "DATEOFJOB1": "date_of_job1",
                         "EXPERIENCE1": "experience_1",
                         "EXPERIENCE2": "experience_2"})
df
```
